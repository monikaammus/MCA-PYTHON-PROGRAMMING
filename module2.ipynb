# create a Rectangle Class

```python
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0
    def set_dimensions(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)
    

```

```python
rect = Rectangle()
rect.set_dimensions(10, 5)

print("Length:", rect.length)
print("Width:", rect.width)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

# Student Grade Manager 

```python
class Student:
    def __init__(self, name, mark1, mark2, mark3):
        self.name = name
        self.mark1 = mark1
        self.mark2 = mark2
        self.mark3 = mark3

    # Method to calculate total marks
    def total_marks(self):
        return self.mark1 + self.mark2 + self.mark3
    def average_marks(self):
        return self.total_marks() / 3
    
```

```python
student1 = Student("Monika", 85, 90, 80)

print("Student Name:", student1.name)
print("Total Marks:", student1.total_marks())
print("Average Marks:", student1.average_marks())
```

# Temperature converter

```python
class Temperature:
    def __init__(self):
        self.celsius = 0

    # Method to set temperature in Celsius
    def set_celsius(self, celsius):
        self.celsius = celsius

    # Method to convert Celsius to Fahrenheit
    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32
    def to_kelvin(self):
        return self.celsius + 273.15

```

```python
temp = Temperature()
temp.set_celsius(25)

print("Temperature in Celsius:", temp.celsius)
print("Temperature in Fahrenheit:", temp.to_fahrenheit())
print("Temperature in Kelvin:", temp.to_kelvin())
```

# Shopping Cart

```python
class ShoppingCart:
    def __init__(self):
        self.items = []  
    def add_item(self, name, price):
        self.items.append({'name': name, 'price': price})
        print(f"{name} added to the cart.")
    def remove_item(self, name):
        for item in self.items:
            if item['name'] == name:
                self.items.remove(item)
                print(f"{name} removed from the cart.")
                return
        print(f"{name} not found in the cart.")
    def total_price(self):
        total = sum(item['price'] for item in self.items)
        return total
    def show_cart(self):
        if not self.items:
            print("Your cart is empty.")
        else:
            print("\nItems in your cart:")
            for item in self.items:
                print(f" - {item['name']} : ₹{item['price']}")
            print(f"Total: ₹{self.total_price()}")


```

```python
cart = ShoppingCart()
cart.add_item("Apples", 120)
cart.add_item("Milk", 50)
cart.add_item("Bread", 40)

cart.show_cart()

cart.remove_item("Milk")
cart.show_cart()
```

# Employee Management

```python
class Employee:
    def __init__(self):
        self.name = ""
        self.emp_id = ""
        self.salary = 0.0

    # Method to set employee details
    def set_details(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    # Method to give a raise by a given percentage
    def give_raise(self, percent):
        self.salary *= (1 + percent / 100)

    # Method to display employee details
    def display_details(self):
        print(f"Employee Name: {self.name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Salary: ₹{self.salary:.2f}")

```

```python
emp = Employee()
emp.set_details("Monika", "E101", 50000)

print("Before Raise:")
emp.display_details()

emp.give_raise(10)

print("\nAfter 10% Raise:")
emp.display_details()
```
