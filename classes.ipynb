```python
class Dog:
    def __init__(self,breed,color,age):
        self.breed=breed
        self.color=color
        self.age=age
    def show(self):
        print('the dog details')
        print('breed self-breed')
        print('age self-age')
```

```python
bruno = Dog('Golden retriever','golden',3)
buddy = Dog('Dalmatian','black&white',5)
sammy = Dog('Bull dog','black',6)
alex = Dog('German sheapard','brown',4)

bruno.show()
buddy.show()
sammy.show()
alex.show()
    
```

```python
doglist=[bruno,sammy,alex,buddy]

#display details of the eldest dogs using logic
age=0
eldest_dog = None
for dog in doglist:
    if age<=dog.age:
        age=dog.age
        eldest_dog=dog
print(eldest_dog.show())
```

```python
class Vehicle:
    def __init__(self,brand):
        self.brand=brand
    def start(self):
        print('Vehicle started')
    def stop(self):
        print('Vehicle stopped')
class Car(Vehicle):
    def start(self):
        print('the car is started')
    def stop(self):
        print('the car is stopped')
vh=Vehicle('abc')
vh.start() 
vh.stop()
car1=Car('honda')
car1.start()
car1.stop()
                  
```

```python

class SI:
    def _init_(self,p,r,t):
        self.p=p
        self.r=r
        self.t=t
    def calc(self):
        ans=self.p+self.t*self.r/100
        return self.p+ans
class CI(SI):
    def calc(self):
        a=self.p*(1+self.r/100)**self.t
        return a    
```

```python
# operator overloading

class Player:
    def __init__(self,Name,Score):
        self.name=Name
        self.score=Score
    def __gt__(self,other):
        return self.score>other.score    
```
