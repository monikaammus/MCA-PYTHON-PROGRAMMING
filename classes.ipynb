```python
class Dog:
    def __init__(self,breed,color,age):
        self.breed=breed
        self.color=color
        self.age=age
    def show(self):
        print('the dog details')
        print('breed self-breed')
        print('age self-age')
```

```python
bruno = Dog('Golden retriever','golden',3)
buddy = Dog('Dalmatian','black&white',5)
sammy = Dog('Bull dog','black',6)
alex = Dog('German sheapard','brown',4)

bruno.show()
buddy.show()
sammy.show()
alex.show()
    
```

```python
doglist=[bruno,sammy,alex,buddy]

#display details of the eldest dogs using logic
age=0
eldest_dog = None
for dog in doglist:
    if age<=dog.age:
        age=dog.age
        eldest_dog=dog
print(eldest_dog.show())
```
